```python
def create_new_budget():
    new_budget = {}
    new_budget["date"] = {}
    new_budget["item"] = {}
    new_budget["cost"] = {}
    
    return new_budget
```

```python
def view_budget(budget):
    print(pd.DataFrame(budget))
```

```python
from datetime import datetime as dt

def add_ele_budget(budget):
    # calculate the index position of new entry in current budget
    idx_pos = -1
    
    for v in budget.values():
        if bool(v) == False:
            idx_pos = 0
    
    if idx_pos < 0:
        df = pd.DataFrame(budget)
        idx_pos = df.index[-1] + 1
    
    # get input from user
    while True:
        date = input("Enter the date of the purchase (yyyy-mm-dd): ")
        try: 
            budget["date"][idx_pos] = dt.strptime(date, "%Y-%m-%d")
            break
        except: 
            print(f"Your input: {date} is in an incorrect format. Please try again.")    
    item = str(input("Enter the item that was purchased: "))
    cost = pyip.inputFloat(prompt="Enter the cost of the item: ")
    
    budget["item"][idx_pos] = item
    budget["cost"][idx_pos] = cost
    
    print(f"Your item: {item} has been sucessfully added to the expense report.")
```

```python
import pyinputplus as pyip
# sort expenses by date or cost(ascending/descending)
def sort_budget(budget):
    
    # takes input from user for sort function
    column_name = pyip.inputChoice(prompt="Which column would you like to sort by (date, cost): ", choices=["date", "cost"])
    order = pyip.inputBool(prompt="Type true for ascending and false for descending: ")    
    
    # converts dictionary to pandas dataframe
    budget = pd.DataFrame(budget)
    
    # sorts based on user input parameters and prints output
    print(budget.sort_values(by=[column_name], axis=0, ascending=order))
    
    
```

```python
# create a print statement after the data is returned in the function
import math
        
def delete_item(budget):
    
    # deleted items held here in dictionary
    deleted_items = create_new_budget()
    deleted_items_df = pd.DataFrame(deleted_items)
    
    # convert dictionary format to pandas dataframe
    df = pd.DataFrame(budget)
    
    while True:
        print(f"The folowing expense report has index values from: 0 to {df.index[-1]}\n")
        index_response = pyip.inputYesNo(prompt="Do you know the index of the line item you'd like to delete (yes/no)?: \n")
        if index_response == "yes".lower() or index_response == "y".lower():
        
            idx_pos_delete = pyip.inputInt(prompt="Please enter the index position here of the item you'd like to delete: \n")
            print("Is this the line item that you would like to delete (yes/no)?: \n", df.iloc[idx_pos_delete])
            validate_delete_item = pyip.inputYesNo()
            if validate_delete_item == "yes".lower() or validate_delete_item == "y".lower():
                
                # save the deleted item to the deleted items dict
                deleted_row = df.iloc[idx_pos_delete]
                deleted_items_dict = deleted_items_df.append(deleted_row, ignore_index=True).to_dict()
                
                # deletes the line item
                df = df.drop(index=idx_pos_delete)
                # resets the index, so it remains sequential
                df = df.reset_index(drop=True)
                print(f"Line {idx_pos_delete} has been deleted.\n")
                
                # return the pandas dataframe converted into a dict object to save the changed data
                return df.to_dict(), deleted_items_dict
            elif validate_delete_item == "no".lower() or validate_delete_item == "n".lower():
                print("No problem, let's try again.")
                continue
        
        elif index_response == "no".lower() or index_response == "n".lower():
            
            if len(df) > 10:
                entry_period = pyip.inputChoice(prompt="Was this a recent entry or an old entry: (recent/old): \n",
                                               choices=["recent", "old"])
                # split the dataframe in half to be able to search either halves based on whether entry was recent or old
                df_length = len(df) / 2

                if entry_period == "old".lower():
                    while df_length > 1:
                        print(df.loc[:df_length])
                        index_found = pyip.inputYesNo(prompt="Do you see your entry (yes/no)?: \n")
                        if index_found == "yes".lower() or index_found == "y".lower():
                            idx_pos_delete = pyip.inputInt(prompt="Please enter the index position here of the item you'd like to delete: \n")
                            # need to put the prompt up here
                            print("Is this the line item that you would line to delete (yes/no)?: \n")
                            print(df.iloc[idx_pos_delete])
                            validate_delete_item = pyip.inputYesNo()

                            if validate_delete_item == "yes".lower() or validate_delete_item == "y".lower():
                                
                                # save the deleted item to the deleted items dict
                                deleted_row = df.iloc[idx_pos_delete]
                                deleted_items_dict = deleted_items_df.append(deleted_row, ignore_index=True).to_dict()
            
                                # deletes the line item
                                df = df.drop(index=idx_pos_delete)
                                # resets the index, so it remains sequential
                                df = df.reset_index(drop=True)
                                print(f"Line {idx_pos_delete} has been deleted.\n")

                                # return the pandas dataframe converted into a dict object to save the changed data
                                return df.to_dict(), deleted_items_dict
                            elif validate_delete_item == "no".lower() or validate_delete_item == "n".lower():
                                print("No problem, let's try again.\n")
                                continue

                        elif index_found == "no".lower() or index_found == "n".lower():
                            df_length = math.ceil(df_length / 2)
                            continue
                    print("I'm sorry, but we couldn't find the entry you were looking for.\nMaybe try looking the the recent entry section.\nLet's try again.\n")

                elif entry_period == "recent".lower():
                    while df_length < len(df)-1:    
                        print(df.loc[df_length:])
                        index_found = pyip.inputYesNo(prompt="Do you see your entry (yes/no)?: \n")
                        if index_found == "yes".lower() or index_found == "y".lower():
                            idx_pos_delete = pyip.inputInt(prompt="Please enter the index position here of the item you'd like to delete: \n")
                            print("Is this the line item that you would line to delete (yes/no)?: \n")
                            print(df.iloc[idx_pos_delete])
                            validate_delete_item = pyip.inputYesNo()
                            if validate_delete_item == "yes".lower() or validate_delete_item == "y".lower():
                                                                
                                # save the deleted item to the deleted items dict
                                deleted_row = df.iloc[idx_pos_delete]
                                deleted_items_dict = deleted_items_df.append(deleted_row, ignore_index=True).to_dict()
                                
                                # deletes the line item
                                df = df.drop(index=idx_pos_delete)
                                # resets the index, so it remains sequential
                                df = df.reset_index(drop=True)
                                print(f"Line {idx_pos_delete} has been deleted.\n")

                                # return the pandas dataframe converted into a dict object to save the changed data
                                return df.to_dict(), deleted_items_dict
                            elif validate_delete_item == "no".lower() or validate_delete_item == "n".lower():
                                print("No problem, let's try again.\n")
                                continue

                        elif index_found == "no".lower() or index_found == "n".lower():

                            # new length thing
                            df_length += ((len(df) - df_length)/2)
                            continue
                    print("I'm sorry, but we couldn't find the entry you were looking for.\nMaybe try looking the the old entry section.\nLet's try again.\n")

            else:
                print(df, "\n")
                idx_pos_delete = pyip.inputInt(prompt="Please enter the index position here of the item you'd like to delete: \n")
                print("Is this the line item that you would like to delete (yes/no)?: \n", df.iloc[idx_pos_delete])
                validate_delete_item = pyip.inputYesNo()
                if validate_delete_item == "yes".lower() or validate_delete_item == "y".lower():
                    
                    # save the deleted item to the deleted items dict
                    deleted_row = df.iloc[idx_pos_delete]
                    deleted_items_dict = deleted_items_df.append(deleted_row, ignore_index=True).to_dict()
                    
                    # deletes the line item
                    df = df.drop(index=idx_pos_delete)
                    # resets the index, so it remains sequential
                    df = df.reset_index(drop=True)
                    print(f"Line {idx_pos_delete} has been deleted.\n")
                
                    # return the pandas dataframe converted into a dict object to save the changed data
                    return df.to_dict(), deleted_items_dict
                elif validate_delete_item == "no".lower() or validate_delete_item == "n".lower():
                    print("No problem, let's try again.")
                    continue

```
